```python
import csv
import numpy as np

bm_time=[]
bm_b=[]
bm_m=[]
with open("line_follow_log_20201120-163149.txt ") as f:
    reader=csv.reader(f,delimiter=' ')
    for row in reader:
        bm_time.append(float(row[0]))
        bm_b.append(float(row[1])/100)
        bm_m.append(float(row[2]))
f.close()
bm_time=np.array(bm_time)
bm_b=np.array(bm_b)
bm_m=np.array(bm_m)

action_time=[]
action_fb=[]
action_skew=[]
action_turn=[]
with open("line_follow_action_log_20201120-163149.txt") as f:
    reader=csv.reader(f,delimiter=' ')
    for row in reader:
        action_time.append(float(row[0]))
        action_fb.append(10*float(row[1])*float(row[4]))
        action_skew.append(10*float(row[2])*float(row[4]))
        action_turn.append(10*float(row[3])*float(row[4]))
f.close()

action_time=np.array(action_time)
action_fb=np.array(action_fb)
action_skew=np.array(action_skew)
action_turn=np.array(action_turn)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import csv
import random
import numpy as np

skew_act_pred_model=nn.Sequential(nn.Linear(4,4),nn.Sigmoid(),nn.Linear(4,1))
#skew_bm_pred_model=nn.Sequential(nn.Linear(3,3),nn.Sigmoid(),nn.Linear(3,2))
skew_bm_pred_model=nn.Linear(3,2)
turn_act_pred_model=nn.Sequential(nn.Linear(4,4),nn.Sigmoid(),nn.Linear(4,1))
#turn_bm_pred_model=nn.Sequential(nn.Linear(3,3),nn.Sigmoid(),nn.Linear(3,2))
turn_bm_pred_model=nn.Linear(3,2)
fb_act_pred_model=nn.Sequential(nn.Linear(4,4),nn.Sigmoid(),nn.Linear(4,1))
fb_bm_pred_model=nn.Sequential(nn.Linear(3,3),nn.Sigmoid(),nn.Linear(3,2))
#fb_bm_pred_model=nn.Linear(3,2)
```

```python
skew_act_pred_x=[]
skew_act_pred_y=[]
for i in range(len(skew_elem[0])):
    skew_act_pred_x.append(torch.tensor([db[skew_elem][i],dm[skew_elem][i],am_m[skew_elem][i],am_b[skew_elem][i]]))
    skew_act_pred_y.append(torch.tensor([action_skew[skew_elem][i]]))
skew_act_pred_x_t=torch.stack(skew_act_pred_x).float()
skew_act_pred_y_t=torch.stack(skew_act_pred_y).float() 


skew_train_x=[]
skew_train_y=[]
for i in range(len(skew_elem[0])):
    skew_train_x.append(torch.tensor([am_m[skew_elem][i],am_b[skew_elem][i],action_skew[skew_elem][i]]))
    skew_train_y.append(torch.tensor([dm[skew_elem][i],db[skew_elem][i]]))
skew_train_x_t=torch.stack(skew_train_x).float()
skew_train_y_t=torch.stack(skew_train_y).float()



print("skew act pred x size {}".format(skew_act_pred_x_t.size()))
print("skew act pred y size {}".format(skew_act_pred_y_t.size()))
```

```python
#train action predictor

def train_act_pred(relevant_elems,action_array,model,name='unknown',lr=0.01):
    #prep the data
    x=[]
    y=[]
    for i in range(len(relevant_elems[0])):
        x.append(torch.tensor([db[relevant_elems][i],dm[relevant_elems][i],am_m[relevant_elems][i],am_b[relevant_elems][i]]))
        y.append(torch.tensor([action_array[relevant_elems][i]]))
    x=torch.stack(x).float()
    y=torch.stack(y).float() 

    loss_fn=nn.MSELoss()
    optimizer = torch.optim.Adagrad(model.parameters(),lr=lr)

    entry=[]
    losslog=[]
    for i in range(30000):
        optimizer.zero_grad()
        y_pred=model(x)
        loss=loss_fn(y_pred,y)
        if i%1000==0 and i !=0:
    #if i%1==0:
            print("Trial {} Loss {}".format(i,loss.item()/len(relevant_elems[0])))
        if i%100==0 and i !=0:
            entry.append(i)
            losslog.append(loss.item()/len(relevant_elems[0]))
        loss.backward()
        optimizer.step()

    fig, ax = plt.subplots()
    ax.plot(entry,losslog)
    ax.set(xlabel='Step', ylabel='Loss',title='')
    fig.show()
    
    y_pred=model(x)
    fig, ax = plt.subplots()
    ax.plot(y,x[:,0],'*')
    ax.plot(y_pred.detach(),x[:,0],'*')
    ax.set(xlabel=name, ylabel='db',title='')
    fig.show()

    fig, ax = plt.subplots()
    ax.plot(y,x[:,1],'*')
    ax.plot(y_pred.detach(),x[:,1],'*')
    ax.set(xlabel=name, ylabel='dm',title='')
    fig.show()

```

```python
train_act_pred(skew_elem,action_skew,skew_act_pred_model,'skew',0.01)
train_act_pred(turn_elem,action_turn,turn_act_pred_model,'turn',0.01)
train_act_pred(fb_elem,action_fb,fb_act_pred_model,'fb',0.01)
```

```python
#train future predictor

def train_future_predictor(relevant_elems,action_array,model,name='unknown',lr=0.01):
    
    x=[]
    y=[]
    for i in range(len(relevant_elems[0])):
        x.append(torch.tensor([am_b[relevant_elems][i],am_m[relevant_elems][i],action_array[relevant_elems][i]]))
        y.append(torch.tensor([db[relevant_elems][i],dm[relevant_elems][i]]))
        
    x=torch.stack(x).float()
    y=torch.stack(y).float()     
    
    loss_fn=nn.MSELoss()
    optimizer = torch.optim.Adagrad(model.parameters(),lr=lr)

    entry=[]
    losslog=[]
    for i in range(150000):
        optimizer.zero_grad()
        y_pred=model(x)
        loss=loss_fn(y_pred,y)
        if i%1000==0 and i !=0:    
            print("Trial {} Loss {}".format(i,loss.item()/len(relevant_elems[0])))
        if i%100==0 and i !=0:    
            entry.append(i)
            losslog.append(loss.item()/len(relevant_elems[0]))    
        loss.backward()
        optimizer.step()

    fig, ax = plt.subplots()
    ax.plot(entry,losslog)
    ax.set(xlabel='Step', ylabel='Loss',title='')
    fig.show()
                               
    y_pred=model(x)
    fig, ax = plt.subplots()
    ax.plot(x[:,2],y[:,0],'*')
    ax.plot(x[:,2],y_pred.detach()[:,0],'*')
    ax.set(xlabel=name, ylabel='db',title='')
    fig.show()

    fig, ax = plt.subplots()
    ax.plot(x[:,2],y[:,1],'*')
    ax.plot(x[:,2],y_pred.detach()[:,1],'*')
    ax.set(xlabel=name, ylabel='db',title='')
    fig.show()
```

```python
train_future_predictor(skew_elem,action_skew,skew_bm_pred_model,name='skew',lr=0.01)
train_future_predictor(turn_elem,action_turn,turn_bm_pred_model,name='turn',lr=0.01)
train_future_predictor(fb_elem,action_fb,fb_bm_pred_model,name='fb',lr=0.01)
```

```python
def test_predictions(act_pred_model,next_pred_model):
    xs_list=[]
    for m in np.linspace(-0.5,0.5,10):
        for b in np.linspace(-0.5,0.5,10):
            for act in np.linspace(-0.5,0.5,10):
                xs_list.append(torch.tensor([b,m,act]))
    xs=torch.stack(xs_list).float()
    ys=next_pred_model(xs)
    ys2_list=[]
    for i in range(len(ys)):
        ys2_list.append(torch.tensor([ys[i][0].item()-xs_list[i][0],ys[i][1].item()-xs_list[i][1],xs_list[i][1],xs_list[i][2]]))
    ys2=torch.stack(ys2_list).float()
    zs=act_pred_model(ys2)
    err=[]
    acts=[]
    for i in range(len(zs)):
        act=xs_list[i][2]
        err.append(zs[i].item()-act)
        acts.append(act)
        
    fig, ax = plt.subplots()
    ax.plot(acts,err,'*')    
    ax.set(xlabel='act', ylabel='err',title='')
    fig.show()
    
test_predictions(skew_act_pred_model,skew_bm_pred_model)
test_predictions(turn_act_pred_model,turn_bm_pred_model)
test_predictions(fb_act_pred_model,fb_bm_pred_model)
        
                
```
