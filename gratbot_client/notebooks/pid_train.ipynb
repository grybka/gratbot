```python
import numpy as np
import csv
import json
from matplotlib import pyplot as plt

data=[]
row_names=["pid_p","pid_i","pid_d","last_prop","last_inte","last_diff","last_lminusr","last_xvel","error_x"]
with open('../to_study/tracker_data_log.txt', 'r') as f:
    for jsonObj in f:        
        example = json.loads(jsonObj)
        data.append(np.array(example))
        
#data=np.array(data)

lengths=[ len(x) for x in data ]
maxlen=np.max(lengths)
plt.figure()
plt.hist(lengths,maxlen)
plt.show()
```

```python
for i in range(len(data)):
    plt.figure()
    plt.title("error")
    plt.plot(data[i][:,8],label="err")
    plt.plot(100*data[i][:,6],label="lr")
    plt.plot(data[i][:,7],label="vel")
    plt.legend()
    #plt.plot(data[0][:,3])
    plt.show()
    #plt.figure()
    #plt.plot(data[0][:,6])
    #plt.show()
```

```python
def prep_dataset_simplest():
    previous_error=[]
    next_error=[]
    p_value=[]
    lminur_value=[]
    for d in data:
        for i in range(len(d)):
            previous_error.append(d[i][3])
            next_error.append(d[i][8])
            p_value.append(d[i][0])
            lminur_value.append(d[i][6])
    return np.array(previous_error),np.array(next_error),np.array(p_value),np.array(lminur_value)

pre,nex,pval,lrval=prep_dataset_simplest()
        
plt.figure()
plt.plot(lrval,nex-pre,'*')
plt.show()
```
